# 3. Manejo de datos

El capping se seleccionó porque equilibra bien la preservación de los datos y la mitigación del impacto de los outliers, sin necesidad de eliminar datos ni transformar la escala de las variables de manera que complique la interpretación. Esta técnica nos permite mantener un análisis limpio y coherente mientras preparamos los datos para modelado más avanzado.

```python
# Definir las columnas que se consideran para capping
columns_capped = ['CRIM', 'ZN', 'B', 'MEDV']

# Función para contar los outliers en las columnas seleccionadas
def count_outliers(df, columns):
    outlier_info = {}
    for col in columns:
        lower_bound = df[col].quantile(0.01)
        upper_bound = df[col].quantile(0.99)
        outliers_lower = (df[col] < lower_bound).sum()
        outliers_upper = (df[col] > upper_bound).sum()
        total_outliers = outliers_lower + outliers_upper
        outlier_info[col] = total_outliers
    return outlier_info

# Contar outliers antes del capping
outliers_before_capping = count_outliers(df, columns_capped)
print("Cantidad de Outliers Antes del Capping")
print(outliers_before_capping)
```

```python
def cap_outliers(df, columns):
    for col in columns:
        percentiles = df[col].quantile([0.01, 0.99]).values
        df[col] = df[col].clip(percentiles[0], percentiles[1])
    return df

# Aplicar capping de outliers
columns_to_cap = ['CRIM', 'ZN', 'B', 'MEDV']
df_capped = cap_outliers(df, columns_to_cap)

# Verificar los cambios
df_capped.describe()
```

Después de aplicar el capping, observamos que los valores máximos de las variables afectadas (como **CRIM**, **ZN**, **B**, y **MEDV**) han sido ajustados, reduciendo el impacto de los outliers en estas columnas. Aquí hay algunas observaciones sobre los cambios:

**CRIM:** El valor máximo se redujo de 88.98 a 41.37, lo que indica que hemos limitado los valores extremadamente altos de criminalidad.

**ZN:** El valor máximo original era 100, ahora es 90, lo cual limita los valores extremadamente altos de proporción de suelo residencial.

**B:** Se ajustó el valor mínimo de 0.32 a 6.73, manteniendo el valor máximo en 396.9.

**MEDV:** El valor mínimo y máximo se ajustaron ligeramente, especialmente en el máximo que pasó de 50 a 50, lo cual puede ser una limitación impuesta por la escala de datos.

**Implicaciones del Capping:**

**Reducción de Sesgo:** Al limitar los valores extremos, el sesgo introducido por estos valores en el modelo se reduce, permitiendo una mejor interpretación de los coeficientes de regresión.

**Preservación de Información:** Mantuvimos todos los registros, simplemente limitando los valores extremos. Esto es útil para análisis de datos en los que deseamos evitar eliminar información potencialmente valiosa.

**Estabilidad del Modelo:** El modelo de regresión será más estable y menos sensible a fluctuaciones extremas en los datos, mejorando así su capacidad de generalización.


```python
# Función para calcular el porcentaje de datos "capped"
def calculate_capped_percentage(original_df, capped_df, columns):
    capped_info = {}
    for col in columns:
        capped_values = (original_df[col] != capped_df[col]).sum()
        capped_percentage = (capped_values / len(original_df)) * 100
        capped_info[col] = capped_percentage
    return capped_info

# Columnas donde aplicamos capping
columns_capped = ['CRIM', 'ZN', 'B', 'MEDV']

# Calcular el porcentaje de datos capped
capped_percentages = calculate_capped_percentage(df, df_capped, columns_capped)
capped_percentages
```

Identificamos y ajustamos los outliers en las variables CRIM, ZN, B y MEDV utilizando capping en los percentiles 1 y 99. Este ajuste fue efectivo, ya que todos los outliers detectados previamente han sido tratados, mejorando así la consistencia y estabilidad del conjunto de datos. Este paso nos asegura que los análisis y modelos posteriores no estarán sesgados por valores extremos.

## Forward Stepwise Regression

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

def forward_selection(data, target):
    initial_features = data.columns.tolist()
    best_features = []
    while initial_features:
        remaining_features = list(set(initial_features) - set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features + [new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if min_p_value < 0.05:  # Considerar solo las variables con p < 0.05
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features

# Forward Stepwise Selection
selected_features = forward_selection(X_train, y_train)
print("Selected features:", selected_features)

# Crear un modelo final usando las características seleccionadas
X_train_selected = sm.add_constant(X_train[selected_features])
model = sm.OLS(y_train, X_train_selected).fit()
y_pred_forward = model.predict(sm.add_constant(X_test[selected_features]))

# Evaluar el rendimiento
forward_rmse = mean_squared_error(y_test, y_pred_forward, squared=False)
forward_r2 = model.rsquared

print(f'Forward Stepwise Regression - RMSE: {forward_rmse}, R²: {forward_r2}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold

# Seleccionar las características elegidas por el método de forward selection
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Configuración de la validación cruzada con 5 pliegues
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Definir el modelo de regresión lineal con las características seleccionadas
model = LinearRegression()

# Aplicar la validación cruzada
cross_val_rmse = cross_val_score(model, X_train_selected, y_train, cv=kf, scoring='neg_root_mean_squared_error')
cross_val_r2 = cross_val_score(model, X_train_selected, y_train, cv=kf, scoring='r2')

# Resultados de la validación cruzada
print(f'Cross-Validated RMSE: {-cross_val_rmse.mean()}, Std Dev of RMSE: {cross_val_rmse.std()}')
print(f'Cross-Validated R²: {cross_val_r2.mean()}, Std Dev of R²: {cross_val_r2.std()}')

# Ajustar el modelo en el conjunto de entrenamiento y predecir en el conjunto de prueba
model.fit(X_train_selected, y_train)
y_pred = model.predict(X_test_selected)

# Evaluar el modelo en el conjunto de prueba
final_rmse = mean_squared_error(y_test, y_pred, squared=False)
final_r2 = model.score(X_test_selected, y_test)

print(f'Final RMSE: {final_rmse}, Final R²: {final_r2}')

```
